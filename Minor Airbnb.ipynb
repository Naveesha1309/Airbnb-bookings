2. pie charts

Business vs. Non-Business Listings: Show the proportion of business and non-business listings using a pie chart.

Room Type Proportions: Display the proportion of each room type using a pie chart.

```python
import matplotlib.pyplot as plt

# Calculate the counts of each category
biz_counts = df['biz'].value_counts()

plt.pie(biz_counts, labels=['Non-business', 'Business'], autopct='%1.1f%%', startangle=60)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

plt.title('Business vs Non-business Distribution')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the counts of each category
biz_counts = df['cleanliness_rating'].value_counts()
labels=df['cleanliness_rating'].unique()

plt.pie(biz_counts, startangle=60)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add a legend based on the unique values
plt.legend(labels, loc="best")

plt.title('Cleanliness rating')

plt.show()
```
